This code defines two functions call_price and put_price that use the Black-Scholes formula to calculate the price of a call option and a put option, respectively. The code then sets the range of the underlying asset price S and uses the Black-Scholes formula to calculate the option prices C(S,t) and P(S,t) for each value of S in the range. Finally, the code plots the option prices C(S,t) and P(S,t) using the matplotlib library.

```python
#Graphing C(S,t) & P(S,t)
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define the call_price and put_price functions using the Black-Scholes formula
def call_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

def put_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

# Set the range of the underlying asset price S
S_min = 50
S_max = 150
S_range = np.linspace(S_min, S_max, 100)

# Set the input parameters for the Black-Scholes formula
S = 100  # Current price of the underlying asset
K = 105  # Option's strike price
T = 1  # Time until the option expires (in years)
r = 0.05  # Risk-free interest rate
sigma = 0.2  # Volatility of the underlying asset


# Calculate the option prices C(S,t) and P(S,t) using the Black-Scholes formula
C = [call_price(S, K, T, r, sigma) for S in S_range]
P = [put_price(S, K, T, r, sigma) for S in S_range]

# Plot the option prices C(S,t) and P(S,t)
plt.plot(S_range, C, label="C(S,t)")
plt.plot(S_range, P, label="P(S,t)")
plt.xlabel("S")
plt.ylabel("Option price")
plt.legend()
```
